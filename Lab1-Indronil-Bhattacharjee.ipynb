**Task 1:** Read in the Default-of-Credit-Card-Clients.csv dataset using functions in Pandas package and print the data in the first two rows.

```python
# Task 1
import pandas as pd

data = pd.read_csv("default-of-credit-card-clients/Default-of-Credit-Card-Clients.csv")
print("Task 1: First two rows of the dataset")
print("="*50)
print(data.head(2))
```

**Task 2:** Print the names and data types of all the columns.

```python
# Task 2
print("Task 2: Column names and data types")
print("="*50)
print(data.dtypes)
```

**Task 3:** Calculate and print the number of rows and columns that this dataset contains. We will
not count the first row because it contains the column names.

```python
# Task 3
num_rows = data.shape[0] - 1  # Excluding the header row with -1
num_cols = data.shape[1]
print("Task 3: Number of rows and columns")
print("="*50)
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

**Task 4:** Calculate and print the distinct values of the column “EDUCATION”.

```python
# Task 4
distinct_values = data["EDUCATION"].unique()
distinct_values_count = data["EDUCATION"].value_counts()
print("Task 4: Distinct values of 'EDUCATION':")
print("="*50)
print("Distinct values:",distinct_values)
print("Distinct value counts:",distinct_values_count)
```

**Task 5:** Calculate and print how many people have “default payment” = 1 and how many people have “default payment” = 0.

```python
# Task 5
default_1_count = data[data["default payment next month"] == 1].shape[0]
default_0_count = data[data["default payment next month"] == 0].shape[0]
print("Task 5: People with default payment counts")
print("="*50)
print("People with default payment = 1:", default_1_count)
print("People with default payment = 0:", default_0_count)
```

**Task 6:** Calculate and print how many people who are married and have “default payment” = 1.

```python
# Task 6
married_default_1_count = data[(data["MARRIAGE"] == 1) & (data["default payment next month"] == 1)].shape[0]
print("Task 6: Married people with default payment = 1")
print("="*50)
print("Married people with default payment = 1:", married_default_1_count)
```

**Task 7:** Calculate and print how many people whose age is greater than 30 and have “default payment” = 1.

```python
# Task 7
age_greater_30_default_1_count = data[(data["AGE"] > 30) & (data["default payment next month"] == 1)].shape[0]
print("Task 7: People with age > 30 and default payment = 1")
print("="*60)
print("People with age > 30 and default payment = 1:", age_greater_30_default_1_count)
```

**Task 8:** Calculate the average value of the “LIMIT_BAL” column when gender is male, and when gender is female.

```python
# Task 8
avg_limit_bal_male = data[data["SEX"] == 1]["LIMIT_BAL"].mean()
avg_limit_bal_female = data[data["SEX"] == 2]["LIMIT_BAL"].mean()
print("Task 8: Average LIMIT_BAL for males and females")
print("="*50)
print("Average LIMIT_BAL for males:", avg_limit_bal_male)
print("Average LIMIT_BAL for females:", avg_limit_bal_female)
```

**Task 9:** Plot a histogram for the column “default payment” when age is less than or equal to 30.
Plot a histogram for the column “default payment” when age is greater than 30.

```python
# Task 9
import matplotlib.pyplot as plt

print("Task 9: Histogram")
print("="*50)
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
data[data["AGE"] <= 30]["default payment next month"].hist()
plt.title("Default Payment for Age <= 30")
plt.subplot(1, 2, 2)
data[data["AGE"] > 30]["default payment next month"].hist()
plt.title("Default Payment for Age > 30")
plt.tight_layout()
plt.show()
```

**Task 10:** Draw a scatter plot with the data of the “AGE” column and the “LIMIT_BAL” column. The x axis represents the “AGE” column and the y axis represents the “LIMIT_BAL” column.

```python
# Task 10
print("Task 10: Scatter Plot")
print("="*50)
plt.figure(figsize=(10, 6))
plt.scatter(data["AGE"], data["LIMIT_BAL"])
plt.xlabel("Age")
plt.ylabel("LIMIT_BAL")
plt.title("Scatter Plot: Age vs LIMIT_BAL")
plt.show()
```
